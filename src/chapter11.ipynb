```python
import pandas as pd
import numpy as np
import random

die = [1, 2, 3, 4, 5, 6]
rolls = pd.DataFrame([(x, y) for x in die for y in die], columns=['Var1', 'Var2'])
rolls['value'] = rolls['Var1'] + rolls['Var2']
print(rolls.head(3))

prob = {'1': 1/8, '2': 1/8, '3': 1/8, '4': 1/8, '5': 1/8, '6': 3/8}
rolls['prob1'] = rolls['Var1'].apply(lambda x: prob[str(x)])
rolls['prob2'] = rolls['Var2'].apply(lambda x: prob[str(x)])
rolls['prob'] = rolls['prob1'] * rolls['prob2']

print(sum(rolls['value'] * rolls['prob']))

wheel = ["DD", "7", "BBB", "BB", "B", "C", "0"]
combos = pd.DataFrame([(x, y, z) for x in wheel for y in wheel for z in wheel], columns=['Var1', 'Var2', 'Var3'])

def get_symbols():
    wheel_probs = [0.03, 0.03, 0.06, 0.1, 0.25, 0.01, 0.52]
    return random.choices(wheel, wheel_probs, k=3)

prob_wheel = {'DD': 0.03, '7': 0.03, 'BBB': 0.06, 'BB': 0.1, 'B': 0.25, 'C': 0.01, '0': 0.52}
combos['prob1'] = combos['Var1'].map(prob_wheel)
combos['prob2'] = combos['Var2'].map(prob_wheel)
combos['prob3'] = combos['Var3'].map(prob_wheel)
combos['prob'] = combos['prob1'] * combos['prob2'] * combos['prob3']

def score(symbols):
    diamonds = symbols.count("DD")
    cherries = symbols.count("C")
    slots = [s for s in symbols if s != "DD"]
    if len(set(slots)) == 1:
        payouts = {'7': 80, 'BBB': 40, 'BB': 25, 'B': 10, 'C': 10, '0': 0}
        return payouts[slots[0]] * (2 ** diamonds)
    elif all(s in ['B', 'BB', 'BBB'] for s in slots):
        return 5 * (2 ** diamonds)
    elif cherries > 0:
        return [0, 2, 5][cherries + diamonds] * (2 ** diamonds)
    else:
        return 0

def play():
    symbols = get_symbols()
    return score(symbols)

combos['prize'] = combos.apply(lambda row: score([row['Var1'], row['Var2'], row['Var3']]), axis=1)

for value in ["My", "first", "for", "loop"]:
    print("one run")

for value in ["My", "second", "for", "loop"]:
    print(value)

chars = [None] * 4
words = ["My", "fourth", "for", "loop"]
for i, word in enumerate(words):
    chars[i] = word
print(chars)

print(sum(combos['prize'] * combos['prob']))

def plays_till_broke(start_with):
    cash = start_with
    n = 0
    while cash > 0:
        cash = cash - 1 + play()
        n += 1
    return n

print(plays_till_broke(100))
```
